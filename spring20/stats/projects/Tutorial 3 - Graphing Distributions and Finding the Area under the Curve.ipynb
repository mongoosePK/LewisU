```python
# This imports a lot of math/special number things we will need.
import numpy as np

# This gives us plotting functionality.
import matplotlib.pyplot as plt
# if using a Jupyter notebook, inlcude:
%matplotlib inline

# This lets us integrate functions. For documentation, see the following link: https://docs.scipy.org/doc/scipy/reference/tutorial/integrate.html
from scipy import integrate

# This will do some special statistics things for us.
import scipy.stats as stats
from scipy.stats import norm
```

### Graphing Functions and Finding the Area under the Curve

```python
def f(x):
    return x

x = np.linspace(0,1,100)
plt.plot(x,f(x))
plt.fill_between(x, f(x), color = "grey",alpha = 0.3, hatch = '|')
```

```python
# This will allow us to integrate and find the area under the curve.
# integrate.quad(function name, lower bound, upper bound)
integrate.quad(f, 0, 1)
# Output gives first the area under the curve, followed by an error estimate.
```

```python
def g(x):
    return x**2


x = np.linspace(0,1,100)
plt.plot(x,g(x))
plt.fill_between(x, g(x), color = "yellow", alpha = 0.2, hatch = '-')
```

```python
integrate.quad(g, 0, 1)
```

#### Recall that in order for a function to be a function for a continuous random variable, it needs to have a total area of 1 under the curve. 

Consider 
$$h(x)=\begin{cases} \frac{3e^{-0.15x}}{20} & x>0, \\ 0 & x \leq 0. \end{cases}$$ 

(Hint: This function should look familiar.)

```python
def h(x):
    return np.exp(-0.15*x) / (20/3)

integrate.quad(h, 0, np.inf) 
```

```python
x = np.arange(0, 50) 
plt.plot(x, h(x))

#This will allow us to shade the area under a specific part of the curve.
x_sub = np.arange(5, 15)
plt.fill_between(x_sub, h(x_sub), color = "red",alpha = 0.3, hatch = '|')
plt.show()

# This will allow us to find the area for the region we shaded. Don't forget to change your bounds if needed!
integrate.quad(h, 5, 15)
```

### The Normal Curve (because it's just weird enough to get it's own libraries)

```python
# This code will help us get the standard normal curve, for our z-scores.

# define constants
mu = 1000 
sigma = 100
x1 = 800
x2 = 1200

# calculate the z-transform
z1 = ( x1 - mu ) / sigma
z2 = ( x2 - mu ) / sigma

x = np.arange(z1, z2, 0.001) # range of x in spec
x_all = np.arange(-10, 10, 0.001) # entire range of x, both in and out of spec
# mean = 0, stddev = 1, since Z-transform was calculated
y = norm.pdf(x,0,1)
y2 = norm.pdf(x_all,0,1)

# build the plot
fig, ax = plt.subplots(figsize=(9,6))
plt.style.use('fivethirtyeight')
ax.plot(x_all,y2)

ax.fill_between(x,y,0, alpha=0.3, color='b')
ax.fill_between(x_all,y2,0, alpha=0.1)
ax.set_xlim([-4,4])
ax.set_xlabel('# of Standard Deviations Outside the Mean')
ax.set_yticklabels([])
ax.set_title('Normal Gaussian Curve')

plt.savefig('normal_curve.png', dpi=72, bbox_inches='tight')
plt.show()
```

```python
# stats.norm.cdf(cutoff value, mean, standard deviation)
area_under_curve = stats.norm.cdf(z2, 0, 1) - stats.norm.cdf(z1, 0, 1)
area_under_curve
```
